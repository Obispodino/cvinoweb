```python
#import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# Load the dataset


# Define the base directory containing data files
base_dir = "/Users/dino/code/Obispodino/cvino/raw_data/X-Wines_Official_Repository/last"  # Updated to absolute path
wines_file = os.path.join(base_dir, "XWines_Full_100K_wines.csv")
wines_data = pd.read_csv(wines_file)
wd = wines_data
wd.head()
```

```python
wd.info()
```

```python
# Get unique regions and grapes
unique_regions = sorted(wd['RegionName'].unique())

# Get only the 59 most frequent single grape varieties (excluding blends)
def is_single_grape(grape_str):
    if pd.isna(grape_str):
        return False
    grape_str = str(grape_str).strip()
    # Check for common blend indicators
    if ',' in grape_str or ' and ' in grape_str or '&' in grape_str or '+' in grape_str or '-' in grape_str:
        return False
    # Check for phrases containing "blend"
    if 'blend' in grape_str.lower() or 'mix' in grape_str.lower():
        return False
    return True

# Filter to only single grape varieties
single_grapes_df = wd[wd['Grapes'].apply(is_single_grape)]
grape_counts = single_grapes_df['Grapes'].value_counts()
unique_grapes = grape_counts.head(59).index.tolist()

# Display summary counts
print(f"Number of unique regions: {len(unique_regions)}")
print(f"Number of unique grape varieties: {len(unique_grapes)}")

# Display first 15 regions
print("\nSample of unique regions:")
for region in unique_regions[:15]:
    print(f"- {region}")

# Display first 15 grape varieties
print("\nSample of unique grape varieties:")
for grape in unique_grapes[:15]:
    print(f"- {grape}")

# Note: Full lists stored in variables 'unique_regions' and 'unique_grapes'
```

```python
# Create a mapping of regions to their countries
try:
    # Load the wine_features.csv file
    features_path = "/Users/dino/code/Obispodino/cvinoweb/wine_features.csv"
    if os.path.exists(features_path):
        wine_features = pd.read_csv(features_path)

        # Check if the required columns exist
        if 'RegionName' in wine_features.columns and 'Country' in wine_features.columns:
            # Create a mapping dictionary from region to country
            region_to_country = {}

            # Get unique region-country pairs
            region_country_pairs = wine_features[['RegionName', 'Country']].drop_duplicates()

            # Create a dictionary for lookup
            for _, row in region_country_pairs.iterrows():
                region_to_country[row['RegionName']] = row['Country']

            # Create a dataframe with regions and countries
            regions_df = pd.DataFrame({'Region': unique_regions})
            regions_df['Country'] = regions_df['Region'].map(region_to_country)

            # Handle regions not found in the mapping
            regions_df['Country'] = regions_df['Country'].fillna('Unknown')

            # Save to CSV
            output_path = "/Users/dino/code/Obispodino/cvinoweb/notebooks/unique_regions.csv"
            regions_df.to_csv(output_path, index=False)
            print(f"Successfully created {output_path}")
            print(f"Found countries for {regions_df['Country'].ne('Unknown').sum()} out of {len(regions_df)} regions")

            # Display the first few rows
            regions_df.head(10)
        else:
            print("Required columns (RegionName, Country) not found in wine_features.csv")
    else:
        # If wine_features.csv doesn't exist, try to use the existing dataset
        if 'Country' in wd.columns:
            # Create a mapping using the wine dataset we already have loaded
            region_country_df = wd[['RegionName', 'Country']].drop_duplicates()

            # Create a dataframe with unique regions
            regions_df = pd.DataFrame({'Region': unique_regions})

            # Merge with the mapping
            regions_df = regions_df.merge(region_country_df,
                                         left_on='Region',
                                         right_on='RegionName',
                                         how='left')

            # Handle regions not found in the mapping
            regions_df['Country'] = regions_df['Country'].fillna('Unknown')

            # Select and reorder columns
            regions_df = regions_df[['Region', 'Country']]

            # Save to CSV
            output_path = "/Users/dino/code/Obispodino/cvinoweb/notebooks/unique_regions.csv"
            regions_df.to_csv(output_path, index=False)
            print(f"Successfully created {output_path} using the wine dataset")
            print(f"Found countries for {regions_df['Country'].ne('Unknown').sum()} out of {len(regions_df)} regions")

            # Display the first few rows
            regions_df.head(10)
        else:
            print("Wine features file not found and 'Country' column not available in the loaded dataset")
            print("Creating CSV with only regions")
            regions_df = pd.DataFrame({'Region': unique_regions, 'Country': 'Unknown'})
            regions_df.to_csv("/Users/dino/code/Obispodino/cvinoweb/notebooks/unique_regions.csv", index=False)
except Exception as e:
    print(f"Error creating region-country mapping: {e}")
```
